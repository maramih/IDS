```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("student-por.csv", delimiter=";")
data.head()
```

```python
data.columns
```

```python
data['absences'].plot(figsize=(20, 5))
```

```python
data.groupby(['reason'])['absences'].sum().sort_values().plot.bar()
```

```python
data.groupby(['reason'])['failures'].sum().plot.pie()
```

```python
d = data.groupby('age')['absences'].sum().sort_values(ascending=False)
d.plot(kind='bar')
```

```python
data.groupby(['sex'])['absences'].sum().plot(kind='bar')
```

```python
data.groupby(['age', 'famsize'])['absences'].sum().plot(kind='bar', figsize=(20, 10))
```

```python
data.groupby('absences')['failures'].sum().plot.pie(figsize=(10, 10))
```

```python
data.groupby(['famsize'])['absences'].sum().plot.pie(figsize=(10, 10))
```

```python
data.groupby('age')[['freetime', 'studytime', 'traveltime']].sum().plot(stacked ='False', figsize=(10,10))
```

```python
data.groupby('absences')[['freetime', 'studytime', 'traveltime']].sum().plot.bar(figsize=(20,10))
```

```python
data.groupby('famsize')[['freetime', 'studytime', 'traveltime']].sum().plot.area(figsize=(20,10))
```

```python
data.groupby('activities')['failures'].sum().plot.bar(figsize=(10,15))
```

```python
data.groupby(['Mjob'])['absences'].sum().plot.bar(figsize=(10, 10))
```

```python
data.groupby(['Fjob'])['absences'].sum().plot.bar(figsize=(10, 10))
```

```python
data.groupby(['Mjob', 'Fjob'])['absences'].sum().plot.bar(figsize=(10, 10))
```

```python
data.groupby(['Mjob'])['failures'].sum().plot.pie(figsize=(10, 10))
```

```python
data.groupby(['Fjob'])['failures'].sum().plot.pie(figsize=(10, 10))
```

```python
data.groupby(['Mjob', 'Fjob'])['failures'].sum().plot.pie(figsize=(10, 10))
```
