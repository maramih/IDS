<h1>LOGISTIC REGRESSION</h1>

   Regresia logistica este o metoda de analiza multivariata a datelor de marketing din ce in ce mai des utilizata datorita faptului ca numarul conditiilor ce trebuie sa fie indeplinite este mai redus decat, de exemplu, in cazul analizei discriminante.
   
   Regresia logistica modeleaza relatia dintre o multime de variabile independente xi (categoriale, continue) si o variabila dependenta dihotomica (nominala, binara) Y. Spre deosebire de regresia clasica, aici una sau mai multe variabile predicitive/explicative (variabile independente) pot fi categoriale, obligatoriu variabila dependenta, deci in cazul acestui nou model este vorba de caracterul nenumeric, calitativ (categorial) al unora dintre variabilele sale.
   
   Modelul regresiv logistic va estima proportia subiectilor care vor avea aceeasi caracteristica de interes pentru stadiul statistic, sau echivalent, probabilitatea ca un subiect oarecare din populatie sa aiba o anumita caracteristica. 
   
   Implementarea regresiei logistice din scikit-learn poate fi accesata din clasa LogisticRegression, iar ea poate face fit binar, Unul vs Restul(One-vs-Rest) sau multinomial, cu doua optiuni de regularizare, cu scopul de a minimiza functia de cost.
   
   Solverii implementati in aceasta clasa sunt: "liblinear", "newton-cg","lbfgs", "sag" si "saga".
   
   *liblinear - utilizeaza algoritmul CD (coordinate descent), insa acesta nu poate invata un adevarat model multinomial, problema fiind optimizata cu ajutorul metodei One vs Rest pentru a putea fi antrenat pe bucati binare pentru toate clasele.
   
   *sag - utilizeaza Stochastin Average Gradient Descent; este mai rapid pentru un set mare de date, cand sunt atat numarul de sample-uri,cat si cel de atribute este destul de mare.
   
   *lbfgs - este un algoritm de optimizare care aproximeaza algoritmul Broyden–Fletcher–Goldfarb–Shanno; recomandat pentru un set redus de date, insa performanta sa are de suferit pentru seturi mai largi.

<h1>RIDGE REGRESSION</h1>

   Ridge Regression este o tehnică pentru analiza datelor de regresie multiple care suferă de multicoliniaritate. Cand
se întâmplă multicoliniaritatea, estimările celor mai mici pătrate sunt imparțiale, dar variațiile lor sunt mari, astfel încât acestea pot fi departe de adevărata lor valoare. Prin adăugarea unui grad de părtinire la estimările de regresie, ridge regression reduce erorile standard. Se speră ca efectul final va fi acela de a oferi estimări mai fiabile. 
   
   Unul dintre principalele obstacole în utilizarea Ridge Regression este alegerea unei valori adecvate de k. Hoerl și Kennard
(1970), inventatorii Ridge Regression, au sugerat utilizarea unui grafic pe care l-au numit Ridge Trace. Acest plot
arată coeficienții de Ridge Regression in funcție de k. La vizualizarea Ridge Trace, analistul alege o valoare
pentru k pentru care coeficienții de regresie s-au stabilizat. Adesea, coeficienții de regresie vor varia foarte mult pentru
valori mici ale k și apoi se vor stabiliza. Alegand cea mai mică valoare a lui k posibil (care introduce cel mai mic bias), coeficienții de regresie par să rămână constanți. Deci creștereac lui k va conduce în cele din urmă
coeficienții de regresie la zero. Un exemplu de Ridge Trace este mai jos:



   Totodata, implementarea Ridge Regression poate fi accesata din clasa Ridge, iar solverii implementati in aceasta sunt:
   
   *'Svd' - folosește o descompunere a valorii singulare a lui X pentru a calcula coeficienții Ridge. Mai stabil pentru matricile singulare decât "cholesky".
   
   *"Cholesky" -  folosește funcția standard scipy.linalg.solve pentru a obține o soluție în formă închisă.
   
   *"Sparse_cg" - utilizează solverul conjugat gradient așa cum se găsește în scipy.sparse.linalg.cg. Ca algoritm iterativ, acest solver este mai adecvat decât "cholesky" pentru date la scară largă (posibilitatea de a seta tol și max_iter).
   
   *'Lsqr' -  utilizează rutina dedicată celui mai mic pătrat regularizat scipy.sparse.linalg.lsqr. Este cea mai rapidă și folosește o procedură iterativă.
   
   *"Sag"-  folosește o coborâre medie a gradientului stochastic, iar "saga" folosește versiunea sa îmbunătățită, imparțială numită SAGA. Ambele metode utilizează o procedură iterativă și sunt adesea mai rapide decât alți solveri atunci când atât n_samples, cât și n_features sunt mari. 
   
   Convergența rapidă "sag" și "saga" este garantată doar pentru caracteristicile cu aproximativ aceeași scară. Puteți pre-prelucra datele cu ajutorul unui scalar din sklearn.preprocessing.
   
   Ultimii cinci solveri susțin atât date complexe, cât și mai simple. Cu toate acestea, numai "sag" și "saga" acceptă intrari mai simple atunci când fit_intercept este Adevărat.

<h1> Bayesian Regression </h1>

<font size="4">Bayesian Ridge estimează un model probabilistic al probleme de regresie. Valoarea pentru parametru w este data de un Gaussian sferic:

Valorile pentru alpha și lambda sunt alese pentru a fi distribuții gamma, conjugatele anterioare pentru precizia Gaussianei.

Modelul rezultat se numește regresie Bayesian Ridge și este similar cu Ridge clasic. Parametrii w, alpha si lambda sunt estimati în comun în timp ce se face fit-ul modelului. Hiperparametrii rămasi sunt parametrii gamma anteriori peste alpha și lambda. Acestea sunt de obicei alese să nu fie informative. Parametrii sunt estimați prin maximizarea probabilității logice marginale.

În comparație cu estimatorul OLS, valorile coeficientului w sunt ușor deplasate către zerouri, ceea ce le stabilizează.

Deoarece precedentele valori ale lui w sunt valori Gaussiene, histograma coeficientilor w estimati este Gaussiana.

Estimarea modelului se face prin maximizarea iterativă a probabilității logice marginale.

De asemenea, comparam predicțiile și incertitudinile pentru regresia Ridge Bayesian pentru o regresie dimensională, utilizând o extindere a caracteristicilor polinomiale. Incertitudinea începe să urce pe partea dreaptă a graficului. Acest lucru se datorează faptului că aceste eșantioane de testare sunt în afara gamei de probe de antrenament.

<h1>PERCEPTRON </h1>

Perceptronul este un alt algoritm simplu, adecvat pentru învățarea pe scară largă. În mod implicit:

Nu necesită o rată de învățare.

Nu este regularizat.

Actualizează modelul doar la greșeli.

Ultima caracteristică implică faptul că perceptronul este mai rapid de antrenat decat SGD cu functia hinge loss,iar modelele rezultate sunt mai dispersate.
